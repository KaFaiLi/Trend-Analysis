# Transition Analysis in Hidden Markov Model: Long-Term and Short-Term Trends

This module analyzes transitions in hidden states generated by a Gaussian HMM to detect suspicious activity. It identifies frequent short-term changes and significant long-term state changes, visualizing these insights for further analysis.

---

## Code Design

The code is structured into the following components:

1. **State Run Identification**:
   - Extracts consecutive runs of the same hidden state using the `get_runs` function.
   
2. **Short-Term and Long-Term Transition Analysis**:
   - Analyzes state transitions for frequent short-term changes using a rolling window.
   - Detects mid-/long-term transitions by identifying extended runs of the same state.

3. **Integration with DataFrame**:
   - Stores analysis results in a DataFrame for alignment with the original data.

4. **Visualization**:
   - Highlights suspicious transitions (short-term frequent changes and mid-/long-term transitions) on time series plots.

5. **Summary Statistics**:
   - Summarizes the frequency and percentage of suspicious activities in the data.

---

## Calculations and Methodology

### Consecutive State Runs
- Identifies runs of the same state to detect long-term trends.
- **Logic**:
  - Traverse the `hidden_states` array.
  - Track start and end indices of runs where the state remains constant.

### Short-Term Frequent Changes
- Calculates state transitions (1 for change, 0 for no change) in the sequence.
- Uses a rolling window to count transitions.
- Flags windows where the fraction of transitions exceeds a specified threshold (`volatility_threshold`).

**Formula for Volatility Detection**:
\[
\text{Volatility Fraction} = \frac{\text{Number of Transitions in Window}}{\text{Window Size}}
\]
- If `Volatility Fraction > volatility_threshold`, the period is flagged as volatile.

### Mid-/Long-Term State Changes
- Runs of the same state exceeding `long_run_threshold` are identified.
- Transition points (end of a run) are marked for further analysis.

### Summary Statistics
- Calculates:
  - Total number of short-term frequent change periods.
  - Total mid-/long-term change points.
  - Percentages of flagged periods in the total data.

---

## Related Statistical Concepts

1. **Rolling Window Analysis**:
   - Used for identifying trends or patterns over short, consecutive subsets of data.

2. **State Transition Analysis**:
   - Tracks transitions between states to detect behavioral anomalies.

3. **Volatility Analysis**:
   - Measures the variability or rate of change in a system over a short period.

4. **Time Series Visualization**:
   - Highlights patterns, anomalies, and transitions for better interpretability.



````python
# Longterm shorterm transition analysis
def get_runs(states):
    """
    Identify consecutive runs of the same state.
    Returns a list of tuples: (state, start_index, end_index).
    """
    runs = []
    start_idx = 0
    current_state = states[0]

    for i in range(1, len(states)):
        if states[i] != current_state:
            runs.append((current_state, start_idx, i - 1))
            current_state = states[i]
            start_idx = i

    # Handle the final run (from start_idx to the last index)
    runs.append((current_state, start_idx, len(states) - 1))
    return runs

def analyze_transitions(hidden_states,
                        window_size=10,
                        volatility_threshold=0.5,
                        long_run_threshold=30):
    """
    Analyze state transitions to detect:
      1) Short-term frequent changes
      2) Mid-/long-term state changes (run >= long_run_threshold)

    Parameters
    ----------
    hidden_states : array-like
        Sequence of hidden states predicted by HMM
    window_size : int
        Size of rolling window for transition analysis (short-term frequent changes)
    volatility_threshold : float
        Threshold (0-1) for flagging high transition windows (fraction of window_size)
    long_run_threshold : int
        Minimum consecutive timesteps in one state before switching is flagged

    Returns
    -------
    dict with:
      - 'high_volatility_mask': boolean array indicating short-term frequent changes
      - 'mid_long_term_change': list of indices where a long run ended and changed states
    """
    # 1) Short-Term Frequent Changes
    # Calculate transitions (1 if state changed, 0 if remained same)
    transitions = (hidden_states[1:] != hidden_states[:-1]).astype(int)

    # Rolling sum of transitions
    rolling_transitions = pd.Series(transitions).rolling(window_size).sum()
    # Flag if transitions exceed threshold
    high_volatility_mask = rolling_transitions > (window_size * volatility_threshold)
    # Align with original length (add False at start and end)
    high_volatility_mask = np.concatenate(
        ([False], high_volatility_mask[:-1], [False])
    )

    # 2) Mid-/Long-Term State Changes
    # Detect runs of the same state
    runs = get_runs(hidden_states)
    mid_long_term_change_indices = []

    for (state, start_idx, end_idx) in runs:
        run_length = end_idx - start_idx + 1
        if run_length >= long_run_threshold:
            # The transition out of this run happens at end_idx + 1 (if it exists)
            if end_idx < len(hidden_states) - 1:
                mid_long_term_change_indices.append(end_idx + 1)

    return {
        'high_volatility_mask': high_volatility_mask,
        'mid_long_term_change': mid_long_term_change_indices
    }


# Apply analyze_transitions
transition_analysis = analyze_transitions(
    hidden_states=hidden_states,
    window_size=10,
    volatility_threshold=0.5,
    long_run_threshold=30
)

# Add analysis results to hmm_df
hmm_df['High_Volatility'] = transition_analysis['high_volatility_mask']
hmm_df['Mid_Long_Term_Change'] = hmm_df.index.isin(transition_analysis['mid_long_term_change'])

# Visualization function for suspicious transitions
def plot_suspicious_transitions(df, dates, column):
    """
    Plot a time series with highlighted suspicious transitions:
      - High Volatility (yellow)
      - Mid-/Long-Term Change (blue stars)
    """
    plt.figure(figsize=(15, 8))

    # Plot original time series
    plt.plot(dates, df[column], color='black', alpha=0.6, label='Original Data')

    # Highlight high volatility periods
    plt.fill_between(
        dates,
        df[column].min(),
        df[column].max(),
        where=df['High_Volatility'],
        color='yellow',
        alpha=0.3,
        label='Short-Term Frequent Changes'
    )

    # Mark mid-/long-term changes (blue stars)
    ml_dates = dates[df['Mid_Long_Term_Change']]
    ml_values = df.loc[df['Mid_Long_Term_Change'], column]
    plt.scatter(
        ml_dates,
        ml_values,
        color='blue',
        marker='*',
        s=100,
        label='Mid-/Long-Term Change'
    )

    plt.title(f'Suspicious Transitions - {column}')
    plt.xlabel('Date')
    plt.ylabel('Value')
    plt.legend()
    plt.xticks(rotation=45)
    plt.tight_layout()

    return plt

# Summarize suspicious activity
def summarize_suspicious_activity(df):
    """
    Generate summary statistics for suspicious transitions.
    """
    summary = {
        'total_high_volatility_periods': df['High_Volatility'].sum(),
        'total_mid_long_term_changes': df['Mid_Long_Term_Change'].sum(),
        'percent_high_volatility': (df['High_Volatility'].sum() / len(df)) * 100,
        'percent_mid_long_term_changes': (df['Mid_Long_Term_Change'].sum() / len(df)) * 100,
    }
    return pd.Series(summary)

# Generate and print summary
summary_stats = summarize_suspicious_activity(hmm_df)
print("\nSuspicious Activity Summary:")
print(summary_stats)

# Plot suspicious transitions for each variable
for column in hmm_df.columns[1:-4]:  # Adjust as necessary
    plot_suspicious_transitions(hmm_df, dates, column)
    plt.show()
````

# State Duration Analysis with Adjusted Logic

This implementation focuses on analyzing the durations of consecutive hidden states predicted by a Hidden Markov Model (HMM). It adjusts the analysis to account for stable states (states with high self-transition probabilities) and identifies suspicious segments based on statistical deviations from expected durations. Additionally, the code generates visualizations to aid in understanding the analysis.

---

## Code Design

The solution is divided into two primary functions and an execution block:

1. **`analyze_state_durations_adjusted`**:
   - Identifies stable states based on a threshold for self-transition probabilities.
   - Computes statistics for durations of each state.
   - Flags suspicious segments for both stable and non-stable states.

2. **`plot_state_duration_analysis_adjusted`**:
   - Visualizes state duration statistics with mean and standard deviation.
   - Highlights suspicious segments on a timeline of hidden states.

3. **Execution Block**:
   - Analyzes state durations using the above functions.
   - Prints the computed statistics and suspicious segments.
   - Generates and displays visualizations.

---

## Calculations and Methodology

### Stable State Identification
- Stable states are identified based on self-transition probabilities exceeding a threshold.
- **Formula**:
  \[
  \text{Stable States} = \{i \ | \ P(i \rightarrow i) \geq \text{threshold\_self\_transition}\}
  \]

### Consecutive State Duration
- Tracks the duration of each hidden state and the corresponding start dates.
- **Implementation**:
  - Iterate through `hidden_states`.
  - Record the duration when the state changes.

### Duration Statistics
- For each state, compute the following statistics:
  - **Mean**:
    \[
    \mu = \frac{\sum_{i=1}^n d_i}{n}
    \]
  - **Standard Deviation**:
    \[
    \sigma = \sqrt{\frac{\sum_{i=1}^n (d_i - \mu)^2}{n}}
    \]
  - Minimum and maximum durations.
  - Count of occurrences.

### Suspicious Segment Identification
- A segment is flagged if:
  - Stable state: Duration is significantly shorter than the mean.
  - Non-stable state: Duration deviates significantly (shorter or longer) from the mean.
- **Deviation Thresholds**:
  - Stable states:
    \[
    \text{Duration} < \mu - 0.5 \cdot \sigma
    \]
  - Non-stable states:
    \[
    \text{Duration} < \mu - 2 \cdot \sigma \ \text{or} \ \text{Duration} > \mu + 2 \cdot \sigma
    \]

### Visualization
- **State Duration Statistics**:
  - Bar plot of mean durations with error bars representing standard deviations.
- **Timeline**:
  - Line plot of hidden states with suspicious segments highlighted in red.

---

## Related Statistical Concepts

1. **Transition Matrix**:
   - Encodes probabilities of transitioning between states.
   - Diagonal values (\(P(i \rightarrow i)\)) indicate self-transition probabilities.

2. **Statistical Outliers**:
   - Outliers are durations that deviate significantly from the mean.
   - Adjust thresholds based on whether the state is stable.

3. **Visualization for Interpretability**:
   - Combines statistical insights with graphical representation for clear interpretation.

---

````python
def analyze_state_durations_adjusted(hidden_states, dates, transition_matrix, threshold_self_transition=0.90):
    """
    Analyzes the durations of consecutive state runs and identifies suspicious segments, adjusted for stable states.

    Args:
        hidden_states: Array of predicted hidden states
        dates: Array of corresponding timestamps
        transition_matrix: Transition matrix of the trained HMM
        threshold_self_transition: Threshold to classify states as 'stable'

    Returns:
        state_duration_stats: Dictionary containing statistics for each state
        suspicious_segments: List of suspicious segments with their details
    """
    # Identify stable states based on self-transition probabilities
    stable_states = [
        state_index for state_index in range(transition_matrix.shape[0])
        if transition_matrix[state_index, state_index] >= threshold_self_transition
    ]
    print("Stable states based on self-transition probability:", stable_states)

    # Get consecutive state durations
    state_durations = []
    state_start_dates = []
    current_state = hidden_states[0]
    current_length = 1
    current_start = dates[0]

    for i in range(1, len(hidden_states)):
        if hidden_states[i] == current_state:
            current_length += 1
        else:
            state_durations.append((current_state, current_length))
            state_start_dates.append(current_start)
            current_state = hidden_states[i]
            current_length = 1
            current_start = dates[i]

    # Add the last segment
    state_durations.append((current_state, current_length))
    state_start_dates.append(current_start)

    # Calculate statistics for each state
    state_duration_stats = {}
    for state in np.unique(hidden_states):
        durations = [d for (st, d) in state_durations if st == state]
        state_duration_stats[state] = {
            "mean": np.mean(durations),
            "std": np.std(durations),
            "min": np.min(durations),
            "max": np.max(durations),
            "count": len(durations)
        }

    # Identify suspicious segments
    suspicious_segments = []
    for i, (state, duration) in enumerate(state_durations):
        stats = state_duration_stats[state]

        # If the state is stable, only flag it for unusually short durations
        if state in stable_states:
            if duration < stats["mean"] - 0.5 * stats["std"]:
                suspicious_segments.append({
                    "state": state,
                    "duration": duration,
                    "start_date": state_start_dates[i],
                    "end_date": state_start_dates[i] + pd.Timedelta(days=duration - 1),
                    "deviation": (duration - stats["mean"]) / stats["std"]
                })
        else:
            # For non-stable states, apply the original logic for both short and long durations
            if (duration < stats["mean"] - 2 * stats["std"]) or (duration > stats["mean"] + 2 * stats["std"]):
                suspicious_segments.append({
                    "state": state,
                    "duration": duration,
                    "start_date": state_start_dates[i],
                    "end_date": state_start_dates[i] + pd.Timedelta(days=duration - 1),
                    "deviation": (duration - stats["mean"]) / stats["std"]
                })

    return state_duration_stats, suspicious_segments


def plot_state_duration_analysis_adjusted(state_duration_stats, suspicious_segments, dates, hidden_states):
    """
    Creates visualizations for state duration analysis, including suspicious segments.

    Args:
        state_duration_stats: Dictionary containing duration statistics for each state
        suspicious_segments: List of suspicious segments with details
        dates: Array of timestamps
        hidden_states: Array of predicted hidden states
    """
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 12))

    # Plot 1: State Duration Statistics
    states = list(state_duration_stats.keys())
    means = [stats["mean"] for stats in state_duration_stats.values()]
    stds = [stats["std"] for stats in state_duration_stats.values()]

    ax1.bar(states, means, yerr=stds, capsize=5, alpha=0.7)
    ax1.set_title("Average State Durations with Standard Deviation")
    ax1.set_xlabel("State")
    ax1.set_ylabel("Duration (days)")
    ax1.grid(True)

    # Plot 2: Timeline with Suspicious Segments Highlighted
    unique_states = np.unique(hidden_states)
    colors = plt.cm.tab10(np.linspace(0, 1, len(unique_states)))

    ax2.plot(dates, hidden_states, 'k-', alpha=0.3, label='State Sequence')

    # Highlight suspicious segments
    for segment in suspicious_segments:
        ax2.axvspan(segment["start_date"], segment["end_date"],
                    color='red', alpha=0.3, label=f"Suspicious: State {segment['state']}")

    ax2.set_title("State Sequence with Suspicious Segments Highlighted")
    ax2.set_xlabel("Date")
    ax2.set_ylabel("State")
    ax2.grid(True)

    # Ensure no duplicate legend entries
    handles, labels = ax2.get_legend_handles_labels()
    by_label = dict(zip(labels, handles))
    ax2.legend(by_label.values(), by_label.keys(), loc='upper right')

    plt.tight_layout()
    return fig


# Analyze state durations with the adjusted logic
state_duration_stats, suspicious_segments = analyze_state_durations_adjusted(
    hidden_states, dates, model.transmat_, threshold_self_transition=0.90
)

# Print statistics and suspicious segments
print("\nState Duration Statistics:")
for state, stats in state_duration_stats.items():
    print(f"\nState {state}:")
    print(f"  Mean duration: {stats['mean']:.2f} days")
    print(f"  Std deviation: {stats['std']:.2f} days")
    print(f"  Range: {stats['min']} to {stats['max']} days")
    print(f"  Number of occurrences: {stats['count']}")

print("\nSuspicious Segments:")
for segment in suspicious_segments:
    print(f"\nState {segment['state']}:")
    print(f"  Duration: {segment['duration']} days")
    print(f"  Period: {segment['start_date'].date()} to {segment['end_date'].date()}")
    print(f"  Deviation: {segment['deviation']:.2f} standard deviations")

# Create visualization
fig = plot_state_duration_analysis_adjusted(state_duration_stats, suspicious_segments, dates, hidden_states)
plt.show()
````

# Likelihood-Based Anomaly Detection with HMM

This implementation extends the use of Gaussian HMM by incorporating likelihood-based anomaly detection. The process identifies anomalies in time series data based on the per-sample log probability (likelihood) under the trained HMM model. It also provides visual and statistical summaries of the detected anomalies.

---

## Code Design

The code is structured into three main components:

1. **Likelihood Calculation and Thresholding**:
   - Computes per-sample log-likelihood scores for the dataset.
   - Defines anomaly thresholds using specific percentiles.

2. **Anomaly Visualization**:
   - Plots log-likelihood scores alongside anomaly thresholds.
   - Highlights anomalies in both log-likelihood and hidden state visualizations.

3. **Anomaly Summary**:
   - Generates summary statistics of anomalies for each threshold.
   - Includes the total number of anomalies, their proportion in the dataset, and their distribution across hidden states.

---

## Calculations and Methodology

### Per-Sample Log Probability
- The `score` method computes the log probability of each observation under the trained HMM.
- **Key Formula**:
  \[
  \text{Log Probability} = \log(P(X | \text{HMM}))
  \]
  Where \(X\) is an individual sample, and \(P(X | \text{HMM})\) is the likelihood.

### Threshold Definition
- Thresholds are calculated using percentiles of log-likelihood scores to capture the lower tail of the distribution, representing unlikely observations.
- **Percentile Formula**:
  \[
  \text{Threshold} = \text{Percentile}(P(X | \text{HMM}))
  \]

### Anomaly Indicators
- Anomalies are flagged if the log-likelihood score is below the threshold for a given percentile:
  \[
  \text{Anomaly} = \begin{cases}
  \text{True}, & \text{if } \text{Log Probability} < \text{Threshold} \\
  \text{False}, & \text{otherwise}
  \end{cases}
  \]

---

## Related Statistical Concepts

1. **Log-Likelihood**:
   - Measures how well the HMM explains an observation.
   - Lower log-likelihood scores suggest observations that deviate from typical patterns.

2. **Percentiles**:
   - Divides the distribution of log-likelihood scores into segments.
   - Lower percentiles (e.g., 1st, 5th) identify the most atypical observations.

3. **Anomaly Detection**:
   - Likelihood-based anomaly detection identifies deviations based on model probabilities.

4. **Hidden State Analysis**:
   - Analyzing hidden states during anomalies helps understand their context and implications.




````python
# Likelihood-based Anomaly
# Calculate per-sample log probability correctly
sample_scores = np.array([model.score(pca_data[[i]]) for i in range(len(pca_data))])

# Calculate likelihood threshold using multiple percentiles for sensitivity analysis
percentiles = [1, 5]
thresholds = {f'{p}th': np.percentile(sample_scores, p) for p in percentiles}

# Create anomaly indicators for each threshold
for percentile, threshold in thresholds.items():
    hmm_df[f'Anomaly_{percentile}'] = sample_scores < threshold

# Step 7: Visualize Anomalies with Likelihood Scores
def plot_anomalies_with_likelihood(df, dates, sample_scores, thresholds):
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 10), height_ratios=[2, 1])

    # Plot 1: Log Likelihood Scores
    ax1.plot(dates, sample_scores, 'b-', label='Log Likelihood Score', alpha=0.6)

    # Add threshold lines
    colors = ['r', 'orange', 'yellow']
    for (label, threshold), color in zip(thresholds.items(), colors):
        ax1.axhline(y=threshold, color=color, linestyle='--',
                   label=f'{label} percentile threshold')
        anomaly_idx = sample_scores < threshold
        ax1.scatter(dates[anomaly_idx],
                   sample_scores[anomaly_idx],
                   color=color,
                   alpha=0.5,
                   s=50,
                   label=f'{label} anomalies')

    ax1.set_title('Log Likelihood Scores with Anomaly Thresholds')
    ax1.set_xlabel('Date')
    ax1.set_ylabel('Log Likelihood Score')
    ax1.legend(loc='lower left')
    ax1.tick_params(axis='x', rotation=45)

    # Plot 2: Hidden States with Anomalies
    hidden_states = df['Hidden State'].values
    unique_states = np.unique(hidden_states)

    # Plot states
    for state in unique_states:
        mask = hidden_states == state
        ax2.scatter(dates[mask],
                   np.zeros_like(dates[mask], dtype=float) + state,
                   label=f'State {state}',
                   alpha=0.6)

    # Overlay anomalies
    for (label, _), color in zip(thresholds.items(), colors):
        anomaly_idx = df[f'Anomaly_{label}'].values
        if anomaly_idx.any():
            ax2.scatter(dates[anomaly_idx],
                       hidden_states[anomaly_idx],
                       color=color,
                       marker='x',
                       s=100,
                       label=f'{label} anomalies')

    ax2.set_title('Hidden States with Anomalies')
    ax2.set_xlabel('Date')
    ax2.set_ylabel('Hidden State')
    ax2.legend(loc='lower left')
    ax2.tick_params(axis='x', rotation=45)

    plt.tight_layout()
    plt.show()

# Step 8: Generate Summary Statistics
def generate_anomaly_summary(df, percentiles):
    summary = pd.DataFrame()

    for p in percentiles:
        anomaly_col = f'Anomaly_{p}th'
        anomaly_periods = df[df[anomaly_col]].copy()

        summary.loc[f'{p}th percentile', 'Total Anomalies'] = len(anomaly_periods)
        summary.loc[f'{p}th percentile', 'Percentage of Data'] = (
            (len(anomaly_periods) / len(df)) * 100
        )

        # State distribution during anomalies
        state_dist = anomaly_periods['Hidden State'].value_counts()
        for state in state_dist.index:
            summary.loc[f'{p}th percentile', f'State {state} Anomalies'] = state_dist[state]

    return summary

# Execute the visualization and summary
plot_anomalies_with_likelihood(hmm_df, dates, sample_scores, thresholds)
anomaly_summary = generate_anomaly_summary(hmm_df, percentiles)
print("\nAnomaly Detection Summary:")
print(anomaly_summary)
````

# Suspicious Score Calculation and Visualization for Time Series Analysis

This segment of the code focuses on identifying and highlighting suspicious behaviors in time series data based on multiple analyses. It introduces a composite "Suspicious Score" derived from various criteria and visualizes these scores over time.

---

## Code Design

The code is divided into two main components:

1. **Final Score Calculation**:
   - Marks segments as suspicious based on predefined criteria.
   - Aggregates multiple binary features into a single composite suspicious score.

2. **Visualization**:
   - Plots time series data, highlighting regions based on the suspicious score with distinct colors for easy interpretation.

---

## Calculations and Methodology

### Suspicious Segment Identification
- Loops through predefined `suspicious_segments` to flag date ranges in the dataset as suspicious.
- Uses boolean masks for efficient row selection.



````python
# Final Score Calculation
# 1) Create a column marking suspicious segments from analyze_state_durations_adjusted.
hmm_df['Suspicious_State_Duration'] = False

# For each suspicious segment, set the rows that fall within that segment's date range to True
for segment in suspicious_segments:
    start_date = segment["start_date"]
    end_date = segment["end_date"]
    # Boolean mask for rows within the segment date range
    mask = (dates >= start_date) & (dates <= end_date)
    hmm_df.loc[mask, 'Suspicious_State_Duration'] = True

# 2) Combine all analyses into a single "Suspicious_Score":
#    - High_Volatility
#    - Mid_Long_Term_Change
#    - Suspicious_State_Duration
#    - Anomaly_1th
#    - Anomaly_5th
hmm_df['Suspicious_Score'] = (
    hmm_df['High_Volatility'].astype(int) +
    hmm_df['Mid_Long_Term_Change'].astype(int) +
    hmm_df['Suspicious_State_Duration'].astype(int) +
    hmm_df['Anomaly_1th'].astype(int) +
    hmm_df['Anomaly_5th'].astype(int)
)
# Check distinct values and their counts in Suspicious_Score
distinct_scores = hmm_df['Suspicious_Score'].value_counts().sort_index()

# Print the results
for score, count in distinct_scores.items():
    print(f"Suspicious Score {score}: {count} occurrences")

# Define a function to plot suspicious scores over the time series
def plot_suspicious_score_only(hmm_df, dates, suspicious_scores):
    """
    Plots the time series with suspicious scores highlighted.
    - Yellow for Low Suspicious (score = 1)
    - Orange for Medium Suspicious (score = 2 or 3)
    - Red for Highly Suspicious (score > 3)
    """
    # Create figure and axes
    n_cols = 2  # Number of columns for subplots
    n_rows = (len(hmm_df.columns) - 3 + n_cols - 1) // n_cols  # Exclude datetime and derived columns
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, n_rows * 3))
    axes = axes.flatten()

    # Define color map for suspicious scores
    suspicious_color_map = {
        'Low Suspicious': ('yellow', suspicious_scores == 1),
        'Medium Suspicious': ('orange', suspicious_scores.isin([2, 3])),
        'Highly Suspicious': ('red', suspicious_scores > 3)
    }

    # Plot variables with suspicious scores
    for i, column in enumerate(hmm_df.columns[1:-6]):  # Adjust to exclude derived columns
        ax = axes[i]
        ax.plot(dates, hmm_df[column], label=column, color='black', linewidth=1)

        # Overlay suspicious scores
        for label, (color, mask) in suspicious_color_map.items():
            ax.fill_between(
                dates,
                hmm_df[column].min(),
                hmm_df[column].max(),
                where=mask,
                color=color,
                alpha=0.4,
                label=label  # Legend entry for each level
            )

        ax.set_title(f"Suspicious Scores: {column}")
        ax.tick_params(axis='x', rotation=45)

    # Remove unused subplots
    for j in range(i + 1, len(axes)):
        fig.delaxes(axes[j])

    # Add legend for suspicious scores
    legend_patches = [
        mpatches.Patch(color=color, label=label) for label, (color, _) in suspicious_color_map.items()
    ]
    fig.legend(handles=legend_patches, loc='upper left', fontsize=10)

    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.suptitle("Suspicious Scores Highlighted on Time Series", fontsize=14)
    plt.show()


# Call the function to plot
plot_suspicious_score_only(hmm_df, dates, hmm_df['Suspicious_Score'])
````
